```python
#import all relevant libraries

import os
import requests
import csv
import json
import pandas as pd
import time

from itertools import zip_longest
import itertools
import operator
```

```python
#this is the url for Turo's "Carculator" api 
url1 = 'https://turo.com/api/carculator/result?country=US&deliveryOffered=true&make={}&model={}&year={}'
```

```python
#open the csv list of all years makes and models of cars that we wanna search Turo for
file = open('year-make-model.csv', 'r', encoding='utf-8')
reader = csv.reader(file, delimiter=',')


#initialize the lists of different fields we wanna collect
Make_List = []
Model_List = []
Year_List = []
Days_Booked_Per_Month = []
Earnings_Per_Booked_Day = []
Estimated_Vehicle_Value = []
Recommended_Daily_Price = []
Trim_Style = []
Style_Label = []
Transmission = []
Trim = [] 
Trim_Label = []
Vehicle_Definition_ID = []

#initialize the different parameters we are gonna use when we request the data from the Turo api. Cant just make a simple call, we need to include the headers in our requests.
cookies = {
}

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:78.0) Gecko/20100101 Firefox/78.0',
    'Accept': '*/*',
    'Accept-Language': 'en-US,en;q=0.5',
    'Referer': 'https://turo.com/us/en/carculator?country=US&deliveryOffered=true&make=Acura&marketAreaId=111&model=ILX&year=2016',
    'Connection': 'keep-alive',
    'Cache-Control': 'max-age=0',
    'TE': 'Trailers',
}

params = (
    ('country', 'US'),
    ('deliveryOffered', 'true'),
    ('make', MAKE),
    ('model', MODEL),
    ('year', YEAR),
)

#loop through all rows of the csv file for each different make model and year of car
for row in reader:
    time.sleep(9.7)
    
    YEAR = row[0]
    MAKE = row[1]
    MODEL = row[2]

    print(MAKE)
    print(MODEL)
    print(YEAR)
    print()
    
    #creates our unique url for whichever row we are on that we will then call to get our relevant data
    url = url1.format(MAKE,MODEL,YEAR)
    print(url)

    #make the call to the API to get our data
    response = requests.get(url, headers=headers, cookies=cookies)

    #append our data to our lists and then print out the data so we can verify its coming in correctly as we go
    try:
        json_data = json.loads(response.text)

        Days_Booked_Per_Month.append(json_data["estimatedDaysBookedPerMonth"])
        Earnings_Per_Booked_Day.append(json_data["estimatedEarningsPerBookedDay"]["amount"])
        Estimated_Vehicle_Value.append(json_data["estimatedVehicleValue"]["amount"])
        Recommended_Daily_Price.append(json_data['recommendedDailyPrice']["amount"])
        Style_Label.append(json_data["trimStyle"]["styleLabel"])
        Transmission.append(json_data["trimStyle"]["transmission"])
        Trim.append(json_data["trimStyle"]["trim"])
        Trim_Label.append(json_data["trimStyle"]["trimLabel"])
        Vehicle_Definition_ID.append(json_data["trimStyle"]["vehicleDefinitionId"])
        
        Make_List.append(MAKE)
        Model_List.append(MODEL)
        Year_List.append(YEAR)

        print()
        print("Estimated Days Booked Per Month")
        print(json_data['estimatedDaysBookedPerMonth'])
        print()
        
        print("Estimated Earnings Per Booked Day")
        print(json_data["estimatedEarningsPerBookedDay"]["amount"])
        print()
        
        print("Estimated Vehicle Value")
        print(json_data["estimatedVehicleValue"]["amount"])
        print()
        
        print("Recommended Daily Price")
        print(json_data['recommendedDailyPrice']["amount"])
        print()
        
        print("Style Label")
        print(json_data["trimStyle"]["styleLabel"])
        print()
        
        print("Transmission")
        print(json_data["trimStyle"]["transmission"])
        print()
        
        print("Trim")
        print(json_data["trimStyle"]["trim"])
        print()
        
        print("Trim Label")
        print(json_data["trimStyle"]["trimLabel"])
        print()
        
        print("Vehicle Definition ID")
        print(json_data["trimStyle"]["vehicleDefinitionId"])
        print()
        
        print()
        print()

    except Exception as e: print(e)

    print()
    print("**********************************")
    print()
```

```python
#Zips together the  lists of data we created above. Matches the first/second/third etc. of each list with the first/second/third etc. of the other lists.
ziplist = itertools.zip_longest(Make_List,Model_List,Year_List,Days_Booked_Per_Month,Earnings_Per_Booked_Day,Estimated_Vehicle_Value,Recommended_Daily_Price,Style_Label,Transmission,Trim,Trim_Label,Vehicle_Definition_ID)


#Turns that zipped list into a Pandas df
ziplist_df = pd.DataFrame(ziplist, columns = ["Make","Model","Year","Days_Booked_Per_Month","Earnings_Per_Booked_Day","Estimated_Vehicle_Value","Recommended_Daily_Price","Style_Label","Transmission","Trim","Trim_Label","Vehicle_Definition_ID"])


#Prints first 10 rows of df for basic error checking
print(ziplist_df.head(10))
#print(ziplist_df)


#Saves df as csv in your current working directory.
ziplist_df.to_csv('listy.csv', index = False)
```
